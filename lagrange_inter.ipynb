```python
N = 10
lin = []
q = linspace(0,1,N)

def lagrange(j,q,x):
    return product([(x-q[i])/(q[j]-q[i]) for i in range(len(q)) if i != j], axis=0)

def Leb(q,x):
    return sum([abs(lagrange(i,q,x)) for i in range(len(q))], axis = 0)


#  The chebyshev minimize the Leb function. Try to find the points minimizing this function,
#  using scipy minimization built in function
    

for i in range(len(q)):
       plot(x,lagrange(i,q,x))
```
